```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merge_df = pd.merge(mouse_metadata , study_results,on="Mouse ID")

# Display the data table for preview
merge_df.head()
```

```python
# Checking the number of mice.

#this creates an array of unique mices
total_mice = merge_df["Mouse ID"].unique()
print(total_mice) 

#use len() method to calculate the amount of unique mice in the array 
total_unique_mice = len(total_mice)
print(f"Total number of mice: {total_unique_mice}")
```

```python
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Using .duplicated() to identify duplicate rows in Mouse ID and Timepoint columns in the DataFrame
# When a mice with same Mouse ID and Timepoint , this command will identify them
#and put them in this DataFrame.

duplicate_mice = merge_df[merge_df.duplicated(["Mouse ID", "Timepoint"])]
duplicate_mice

# The above code will show which Mouse ID has duplicates;
# use .unique() function to identify the Mouse ID 

duplicate_ID = duplicate_mice["Mouse ID"].unique()
duplicate_ID

```

```python
# Optional: Get all the data for the duplicate mouse ID. 
# == 'g989' to create a conditional when locating the elements in
# columns "Mouse ID"
duplicate_rows = merge_df.loc[(merge_df["Mouse ID"]=='g989')]
duplicate_rows
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# Using similar conditional as above, but by creating a DF without 
#duplicate (!= 'g989')
remove_duplicate = merge_df.loc[(merge_df["Mouse ID"]!='g989')]
remove_duplicate
                                  
```

```python
# Checking the number of mice in the clean DataFrame.
number_mice = len(remove_duplicate["Mouse ID"].unique())
number_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

group_drug = remove_duplicate.groupby(['Drug Regimen'])
mean = group_drug['Tumor Volume (mm3)'].mean()
median = group_drug['Tumor Volume (mm3)'].median()
variance = group_drug['Tumor Volume (mm3)'].var()
stdv = group_drug['Tumor Volume (mm3)'].std()
SEM = group_drug['Tumor Volume (mm3)'].sem()

# Put all the series into a DataFrame
summary_statistics = pd.DataFrame({"Mean Tumor Volume" : mean, "Median Tumor Volume" : median , 
                                    "Tumor Volume Variance" : variance, "Tumor Volume Std. Dev." : stdv,
                                    "Tumor Volume Std. Err." : SEM})
summary_statistics
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
aggregation_header = {'Tumor Volume (mm3)':['mean','median','var','std','sem']}
aggregation_summary = remove_duplicate.groupby(["Drug Regimen"]).agg(aggregation_header)
aggregation_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Start by getting the last (greatest) timepoint for each mouse
# last timepoint = 45, in this case, we will use .max() to grab the value
max_timepoint = remove_duplicate[["Mouse ID", "Timepoint"]]
greatest = max_timepoint.groupby(["Mouse ID"])["Timepoint"].max()
greatest
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_tumor = pd.merge(greatest, remove_duplicate, how="left", on=["Mouse ID",
                                                                   "Timepoint"])
final_four_drugs = merge_tumor[merge_tumor["Drug Regimen"].isin(drugs)]
final_four_drugs
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"] 

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in drugs:
    quantitative_list = []
    # Locate the rows which contain mice on each drug and get the tumor volumes
    for index, row in final_four_drugs.iterrows():
        if row["Drug Regimen"] == drug:
            quantitative_list.append(row['Tumor Volume (mm3)'])
    
    # add subset 
    #sort the data from row['Tumor Volume (mm3)'] in ascending order
    quantitative_list.sort()
    #add all the values into the empty list above
    tumor_vol_data.append(quantitative_list)
    #convert the list into dataframe
    quantitative_list = pd.DataFrame(quantitative_list)
    
    # Determine outliers using upper and lower bounds
    # the .quantile() only works with pandas Series and DataFrames
    # that is why we converted the list into a DataFrame
    quartiles = quantitative_list.quantile([.25,.5,.75])
    lowerq =quartiles[0][0.25]
    upperq = quartiles[0][0.75]
    IQR = upperq - lowerq
    upper_bounds = upperq + (1.5*IQR)
    lower_bounds = lowerq - (1.5*IQR)
    
    outliers = [upper_bounds, lower_bounds]
    
    print(f"{drug}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

fig1, ax1 = plt.subplots()
ax1.boxplot(tumor_vol_data, labels=drugs)
ax1.set_ylabel("Final Tumor Volume (mm3)")

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
#["Mouse ID"] == "l509"

mouse_l509 = remove_duplicate.loc[(remove_duplicate["Drug Regimen"] == "Capomulin") & (remove_duplicate["Mouse ID"] == "l509")]
plt.plot(mouse_l509["Timepoint"], mouse_l509["Tumor Volume (mm3)"])
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin treatment of mouse l509")
plt.show()
                                                                                    
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
#average, use .mean()
capomulin_regimen = remove_duplicate.loc[remove_duplicate["Drug Regimen"]=="Capomulin"] 
capomulin_group = capomulin_regimen.groupby(["Mouse ID"])
capomulin_group.head()

average_tumor_vol = capomulin_group["Tumor Volume (mm3)"].mean()
average_tumor_vol

average_mouse_weight = capomulin_group["Weight (g)"].mean()
average_mouse_weight

plt.scatter(average_mouse_weight, average_tumor_vol)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
# y = m x + b

X = capomulin_group["Weight (g)"].mean()
Y = capomulin_group["Tumor Volume (mm3)"].mean()

# Perform linear regression
slope, intercept, r_value, p_value, std_err = st.linregress(X, Y)
#formula (y=mx+b)
linear_regression = slope * X + intercept

#r_value is the correlation coefficient
print(f'The correlation between mouse weight and the average tumor volume is {r_value.round(2)}')
print(f' y = {slope.round(2)}x + {intercept.round(2)}')

# remember to import scipy.stats as st to be able to run linregress command
plt.scatter(average_mouse_weight, average_tumor_vol)
plt.plot(X , linear_regression, label = "Regression Line", color = "r")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()


```
